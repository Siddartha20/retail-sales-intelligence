```python
import pandas as pd

df = pd.read_csv("superstore.csv")



```

```python

# basic checks
print(df.head())
print(df.info())
```

```python

# clean
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
```

```python
# new columns
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

print(df.head())
```

```python
import datetime as dt

latest_date = df['Order Date'].max()

rfm = df.groupby('Customer ID').agg({
    'Order Date': lambda x: (latest_date - x.max()).days,
    'Order ID': 'count',
    'Sales': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

print(rfm.head())
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method="first"), 4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

print(rfm.head())
```

```python
def segment(row):
    if row['RFM_Score'] in ['444','443','434','433']:
        return 'High Value'
    elif row['RFM_Score'] in ['344','343','334']:
        return 'Loyal'
    elif row['RFM_Score'] in ['244','243','234']:
        return 'Potential'
    else:
        return 'At Risk'

rfm['Segment'] = rfm.apply(segment, axis=1)

print(rfm.head())
```

```python
print(rfm['Segment'].value_counts())
```

```python
# Top 5 customers
print(rfm.sort_values(by='Monetary', ascending=False).head())

# Segment revenue contribution
segment_sales = rfm.groupby('Segment')['Monetary'].sum()
print(segment_sales)
```
